<H2><B> Report on Clustering Analysis of Funding Round Dataset</B> </H2>

<B><H3>1.Objective:</H3></B>

The objective of this report is to conduct a clustering analysis on a funding round dataset using the K-Means algorithm. The analysis aims to identify distinct segments or groups within the dataset based on various attributes such as the amount raised, valuation, type of funding round, number of participants, etc. By segmenting the dataset, we intend to gain insights into the patterns and characteristics of different types of funding rounds, which can provide valuable information for investors, startups, and other stakeholders in the venture capital ecosystem.

**Methodology:**

**1.1 Data Preprocessing:**
   - The dataset was cleaned to handle missing values and outliers appropriately.
   - Categorical variables were encoded using suitable techniques such as pandas or label encoding.

**1.2 Feature Selection:**
   - Relevant features for clustering were selected, including 'raised_amount_usd', 'pre_money_valuation_usd', 'post_money_valuation_usd', 'participants', etc.

**1.3 Clustering Algorithm:**
   - K-Means clustering algorithm was chosen as the base model for segmenting the funding round dataset.
   - K-Means is a popular unsupervised learning algorithm that partitions data into 'K' clusters based on similarities in feature space.

**1.4 Determining the Appropriate Number of Segments:**
   - The appropriate number of segments or clusters (K) was determined using the within-cluster sum of squares (SS) score and the Davies-Bouldin score (DB).
   - The SS score measures the compactness of clusters, while the DB score evaluates the separation between clusters.
   - Various values of K were tested, and the optimal value was selected based on the elbow method or silhouette score.

```python
# 2.1. K-Means Clustering
# -----------------------

# 2.1.1. Determine Value of 'K' in K-Means using Elbow Curve & KMeans-Inertia
# ---------------------------------------------------------------------------
'''
KMeans-Inertia : Sum of Squared Distances of Samples to their closest Cluster Center (Centroid), Weighted by the Sample Weights (if provided)
'''
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(df_ppd_round)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

```python
# 2.1.3. K-Means Clustering Model Evaluation [K=2 | K=3 | K=4]
# ------------------------------------------------------

sscore_km_2cluster = sscore(df_ppd_round, km_2cluster_model)
print(f"Silhouette Score for K-means clusters (2 clusters): {sscore_km_2cluster}")
dbscore_km_2cluster = dbscore(df_ppd_round, km_2cluster_model)
print(f"Davies-Bouldin Index for K-means clusters (2 clusters): {dbscore_km_2cluster}")

# 2.1.3. K-Means Clustering Model Evaluation [K=2 | K=3 | K=4]
# ------------------------------------------------------

sscore_km_3cluster = sscore(df_ppd_round, km_3cluster_model)
print(f"Silhouette Score for K-means clusters (3 clusters): {sscore_km_3cluster}")
dbscore_km_3cluster = dbscore(df_ppd_round, km_3cluster_model)
print(f"Davies-Bouldin Index for K-means clusters (3 clusters): {dbscore_km_3cluster}")

# 2.1.3. K-Means Clustering Model Evaluation [K=2 | K=3 | K=4]
# ------------------------------------------------------

sscore_km_4cluster = sscore(df_ppd_round, km_4cluster_model)
print(f"Silhouette Score for K-means clusters (4 clusters): {sscore_km_4cluster}")
dbscore_km_4cluster = dbscore(df_ppd_round, km_4cluster_model)
print(f"Davies-Bouldin Index for K-means clusters (4 clusters): {dbscore_km_4cluster}")
```
